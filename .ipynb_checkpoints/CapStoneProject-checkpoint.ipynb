```python
import pandas as pd;
import numpy as np;
import requests;
from bs4 import BeautifulSoup;
```

Get-retrieve the website content

```python
URL = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M";
r = requests.get(URL);
print(r)
```

Get the object of the BeautifulSoup and read from Wikipedia

```python

soup = BeautifulSoup(r.content, 'html5lib')

#print("table ==",soup.body.table.tbody)
table = soup.body.table.tbody
#print(soup.prettify())
postalCodes = [];
borroughs = [];
neighbourhoods =[];
for row in table.findAll('tr'):
        #print("Current tds ==", row.findAll('td'));
        tdRow = row.findAll('td');
        if(len(tdRow)!=0):
            borough = str(tdRow[1]).replace("<td>","").replace("</td>","")
            na = 'Not assigned';
            boolVal = borough == na;
            print("boolVal::",boolVal);
            if(boolVal):
                print("Oops! n/a borough");
            else:
                print("borrough  == ",borough);
                postalCode = str(tdRow[0]).replace("<td>","").replace("</td>","");
                print("postalCode  == ",postalCode);
                neighbourhood = str(tdRow[2]).replace("<td>","").replace("</td>","");
                print("neigbhourhood  == ",neighbourhood);
```
